```python
import cv2
import shutil
import pandas as pd
```

```python
art = pd.read_csv('../Data/Artist/all_data_info.csv')
art.head()
```

# 원본 이미지에서 깨지지 않은 파일만 선택하여 저장

```python
path = '../Data/Artist/original/org_images/'
dst = '../Data/Artist/original/org_select/'
index = []
for i in art.index:
    try :
        file = art['new_filename'][i]
        img = cv2.imread(path + file, cv2.IMREAD_COLOR)
        h, w, n = img.shape
        index.append(i)
        shutil.copy(path + file, dst + file)

    except :
        pass
    
```

```python
art_select = art.iloc[index].reset_index(drop=True)
art_select.head()
```

```python
art_select.to_csv('../Data/Artist/all_select.csv')
```
