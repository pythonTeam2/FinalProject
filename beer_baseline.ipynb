```python
csv_data = beer1[['ABV', 'Ave Rating', 'Astringency','Body', 'Alcohol', 'Bitter', 'Sweet', 'Sour', 'Salty', 'Fruits','Hoppy', 'Spices', 'Malty']]
csv_label =beer1['Style']
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = train_test_split(
    csv_data, csv_label
)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
nn_clf = KNeighborsClassifier(n_neighbors=3) #n_neighbors=3 :하이퍼 파라미터라고한다. 데이터가 많은때는 줄이고, 적을때는 올린다.
nn_clf.fit(train_data,train_label)
print("KNN train :",nn_clf.score(train_data,train_label))
print("KNN test :",nn_clf.score(test_data,test_label))
print('-' * 30)

# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt_clf =DecisionTreeClassifier()
dt_clf.fit(train_data,train_label)
print("Decision Tree train :", dt_clf.score(train_data,train_label))
print("Decision Tree test :", dt_clf.score(test_data,test_label))
print('-' * 30)

# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(train_data,train_label)
print("Random Forest train :", rf_clf.score(train_data,train_label))
print("Random Forest test :", rf_clf.score(test_data,test_label))
print('-' * 30)

# AdaBoost
from sklearn.ensemble import AdaBoostClassifier
ad_clf =AdaBoostClassifier()
ad_clf.fit(train_data,train_label)
print("AdaBoost train :", ad_clf.score(train_data,train_label))
print("AdaBoost test :", ad_clf.score(test_data,test_label))
print('-' * 30)

# Neural Net
from sklearn.neural_network import MLPClassifier
net_clf = MLPClassifier()
net_clf.fit(train_data,train_label)
print("Neural Net train :", net_clf.score(train_data,train_label))
print("Neural Net test :", net_clf.score(test_data,test_label))
print('-' * 30)

# Naive Bayes 
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(train_data,train_label)
print("Naive Bayes train :", nb_clf.score(train_data,train_label))
print("Naive Bayes test :", nb_clf.score(test_data,test_label))
print('-' * 30)

# QDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda_clf = QuadraticDiscriminantAnalysis()
qda_clf.fit(train_data,train_label)
print("QDA train :", qda_clf.score(train_data,train_label))
print("QDA test :", qda_clf.score(test_data,test_label))
print('-' * 30)
```

> 모든 모델의 train점수가 test의 점수보다 잘 나와서 과대 적합이다.

```python
# Warning 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 그래프 꾸미기
# Style 서식 지정
plt.style.use('ggplot')

# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(25,10))

# x축 눈금 라벨 회전하기
plt.xticks(rotation='vertical')

# y축 눈금 라벨 회전하기
plt.yticks(rotation=75)

# 라벨
plt.title('타입별 맛')
plt.xlabel('타입')
plt.ylabel('맛 평균')

# Marker 표시
plt.plot(s, markersize =10)

# 범례표시
plt.legend(labels = ['Astringency','Bitter','Sweet','Sour','Salty'], loc ='best', fontsize = 15)
plt.show()
```
